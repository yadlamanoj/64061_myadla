**NEURAL NETWORKS**


The IMDB dataset


```python
from tensorflow.keras.datasets import imdb
(train1_data, train1_labels), (test1_data, test_labels) = imdb.load_data(num_words=10000)
```

```python
train1_data[0]
```

```python
train1_labels[0]
```

```python
#vectorizing the data
import numpy as np

def vectorize_sequences(sequences, dimension=10000):
    # Create an all-zero matrix of shape (len(sequences), dimension)
    res = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        res[i, sequence] = 1.  # set specific indices of results[i] to 1s
    return res

# Our vectorized training data
x_train = vectorize_sequences(train1_data)
# Our vectorized test data
x_test = vectorize_sequences(test1_data)
```

```python
x_train[0]
```

```python
#vectorizing the labels
y_train = np.asarray(train1_labels).astype('float32')
y_test = np.asarray(test_labels).astype('float32')
```

**1. One and three hidden layers**

```python
from keras import models
from keras import layers

model_onehidden_layer = models.Sequential()
model_onehidden_layer.add(layers.Dense(32, activation='relu', input_shape=(10000,)))
model_onehidden_layer.add(layers.Dense(1, activation='sigmoid'))

model_onehidden_layer.compile(optimizer="rmsprop",
                               loss="binary_crossentropy",
                               metrics=["accuracy"])

x_val = x_train[:10000]
partial_x_train = x_train[10000:]

y_val = y_train[:10000]
partial_y_train = y_train[10000:]

history_one_hidden_layer = model_onehidden_layer.fit(partial_x_train,
                                                      partial_y_train,
                                                      epochs=20,
                                                      batch_size=512,
                                                      validation_data=(x_val, y_val))
import matplotlib.pyplot as plt
def plot_history(history):
    history_dict = history.history
    loss_values = history_dict["loss"]
    val_loss_values = history_dict["val_loss"]
    epochs = range(1, len(loss_values) + 1)

    plt.plot(epochs, loss_values, "bo", label="Training loss")
    plt.plot(epochs, val_loss_values, "b", label="Validation loss")
    plt.title("Training and validation loss")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    plt.show()

    plt.clf()
    acc = history_dict["accuracy"]
    val_acc = history_dict["val_accuracy"]

    plt.plot(epochs, acc, "bo", label="Training acc")
    plt.plot(epochs, val_acc, "b", label="Validation acc")
    plt.title("Training and validation accuracy")
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.show()


plot_history(history_one_hidden_layer)
model_three_hidden_layer = models.Sequential()
model_three_hidden_layer.add(layers.Dense(32, activation='relu', input_shape=(10000,)))
model_three_hidden_layer.add(layers.Dense(32, activation='relu', input_shape=(10000,)))
model_three_hidden_layer.add(layers.Dense(32, activation='relu', input_shape=(10000,)))
model_three_hidden_layer.add(layers.Dense(1, activation='sigmoid'))

model_three_hidden_layer.compile(optimizer="rmsprop",
                               loss="binary_crossentropy",
                               metrics=["accuracy"])

history_three_hidden_layer = model_three_hidden_layer.fit(partial_x_train,
                                                      partial_y_train,
                                                      epochs=20,
                                                      batch_size=512,
                                                      validation_data=(x_val, y_val))
plot_history(history_three_hidden_layer)


```

**2. Layers with more or fewer hidden units**

- Fewer hidden units(16 units)

```python
model_fewer_hidden_units = models.Sequential()
model_fewer_hidden_units.add(layers.Dense(16, activation='relu', input_shape=(10000,)))
model_fewer_hidden_units.add(layers.Dense(16, activation='relu', input_shape=(10000,)))
model_fewer_hidden_units.add(layers.Dense(1, activation='sigmoid'))

model_fewer_hidden_units.compile(optimizer="rmsprop",
                               loss="binary_crossentropy",
                               metrics=["accuracy"])

history_fewer_hidden_units = model_fewer_hidden_units.fit(partial_x_train,
                                                      partial_y_train,
                                                      epochs=20,
                                                      batch_size=512,
                                                      validation_data=(x_val, y_val))
```

```python
plot_history(history_fewer_hidden_units)
```

- More hidden units (64 units)

```python
model_more_hidden_units = models.Sequential()
model_more_hidden_units.add(layers.Dense(64, activation='relu', input_shape=(10000,)))
model_more_hidden_units.add(layers.Dense(64, activation='relu', input_shape=(10000,)))
model_more_hidden_units.add(layers.Dense(1, activation='sigmoid'))

model_more_hidden_units.compile(optimizer="rmsprop",
                               loss="binary_crossentropy",
                               metrics=["accuracy"])

history_more_hidden_units = model_more_hidden_units.fit(partial_x_train,
                                                      partial_y_train,
                                                      epochs=20,
                                                      batch_size=512,
                                                      validation_data=(x_val, y_val))
```

```python
plot_history(history_more_hidden_units)
```

**3. Using the MSE loss function instead of binary_crossentropy.**

```python
from keras import models
from keras import layers

model_mse_loss = models.Sequential()
model_mse_loss.add(layers.Dense(32, activation='relu', input_shape=(10000,)))
model_mse_loss.add(layers.Dense(32, activation='relu', input_shape=(10000,)))
model_mse_loss.add(layers.Dense(1, activation='sigmoid'))

model_mse_loss.compile(optimizer="rmsprop",
                               loss="mse",
                               metrics=["accuracy"])

history_mse_loss = model_mse_loss.fit(partial_x_train,
                                      partial_y_train,
                                      epochs=20,
                                      batch_size=512,
                                      validation_data=(x_val, y_val))
```

```python
plot_history(history_mse_loss)
```

**4. Using the tanh activation instead of relu.**

```python
model_tanh_activation = models.Sequential()
model_tanh_activation.add(layers.Dense(32, activation='tanh', input_shape=(10000,)))
model_tanh_activation.add(layers.Dense(32, activation='tanh', input_shape=(10000,)))
model_tanh_activation.add(layers.Dense(1, activation='sigmoid'))

model_tanh_activation.compile(optimizer="rmsprop",
                               loss="mse",
                               metrics=["accuracy"])

history_tanh_activation = model_tanh_activation.fit(partial_x_train,
                                      partial_y_train,
                                      epochs=20,
                                      batch_size=512,
                                      validation_data=(x_val, y_val))
```

```python
plot_history(history_tanh_activation)
```

**5. Using Regularization and Dropout.**

```python
from keras import regularizers
model_regularization_dropout = models.Sequential([
    layers.Dense(32, activation="tanh", kernel_regularizer=regularizers.l2(0.001)),
    layers.Dropout(0.5),
    layers.Dense(32, activation="tanh", kernel_regularizer=regularizers.l2(0.001)),
    layers.Dropout(0.5),
    layers.Dense(1, activation="sigmoid")])
model_regularization_dropout.compile(optimizer="rmsprop", loss="binary_crossentropy", metrics=["accuracy"])
history_regularization_dropout = model_regularization_dropout.fit(partial_x_train,
                                                                  partial_y_train,
                                                                  epochs=20,
                                                                  batch_size=512,
                                                                  validation_data=(x_val, y_val))
```

```python
plot_history(history_regularization_dropout )
```

**Summary:**

In this exercise, different neural network configurations were tested on the IMDB dataset, adjusting the number of hidden layers, units, activation functions, and loss functions. Despite incorporating regularization techniques like L2 regularization and Dropout, the model exhibited overfitting after several epochs. This was evident as the validation accuracy plateaued while the validation loss increased.

From the graphs it is observed that--
1.	Fewer Hidden Units (16 units):
- The model reached close to 100% training accuracy by the end of training.
- However, validation accuracy leveled off around 87%, and validation loss increased sharply towards the end, suggesting overfitting. This implies the model performed well on the training data but struggled to generalize to new data.
2.	More Hidden Units (64 units):
- This model also achieved 100% training accuracy.  
- Validation accuracy remained fairly consistent, fluctuating around 87%, but the validation loss increased over time, indicating overfitting, similar to the model with fewer hidden units.
3.	Using MSE Loss Function Instead of Binary Cross-Entropy:
- The model's performance differed slightly in terms of loss values due to the change in the loss function.  
- Validation accuracy was similar to the models using binary cross-entropy, reaching about 87-88%, with validation loss remaining at acceptable levels, though slightly higher than the binary cross-entropy model.
4.	Using Tanh Activation Instead of Relu:
- The model with the tanh activation function achieved strong accuracy, performing similarly to the relu-based models.  
- Validation accuracy and loss were comparable, suggesting that changing the activation function had little effect on performance in this instance.

Overall, models with both fewer and more hidden units performed similarly, showing no significant improvement from increasing the number of hidden units. However, all models experienced overfitting. The model using MSE as the loss function had slightly weaker performance compared to the one with binary cross-entropy, while the model with tanh activation performed on par with the relu-based model.
